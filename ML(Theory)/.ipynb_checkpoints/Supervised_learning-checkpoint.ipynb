# ***Types of Problems in Supervised Learning***

Supervised Learning mainly consists of **two types of problems**:

---

# 1. Regression

Regression is used when the output is a **continuous numerical value**.

### Examples
- House Price Prediction
- Salary Prediction
- Temperature Prediction
- Stock Price Prediction

### Input (X)
- Area
- Number of Bedrooms
- Location

### Output (Y)
- House Price

---

# 2. Classification

Classification is used when the output is a **category or label**.

### Examples
- Spam or Not Spam
- Disease or No Disease
- Cat or Dog
- Positive or Negative Review

### Input (X)
- Email Text
- Sender Email
- Link
- Exclamations

### Output (Y)
- Spam / Not Spam

---

# Comparison

| Regression | Classification |
|------------|----------------|
| Predicts numerical values | Predicts categories or labels |
| Output is continuous | Output is discrete |
| Example: House Price | Example: Spam Detection |

---

# Summary

```
                Supervised Learning
                       |
          --------------------------
          |                        |
     Regression             Classification
          |                        |
 Predicts Numbers          Predicts Categories
 (Continuous Output)        (Discrete Output)
```

---

## Key Points

- Regression predicts continuous numerical values.
- Classification predicts categories or labels.
- These are the two major types of problems in Supervised Learning.
