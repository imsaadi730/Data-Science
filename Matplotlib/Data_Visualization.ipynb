# ***Bar chartsor Bar plots***
- A bar chart or bar plot is used to display categorical data by using rectangular bars.
- Each bar's height represents a value.
- syntex for verticl bar:plt.bar(x,y)
- syntex for horizontal bar:plt.barh(x,y)

```python
movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years,revenue,color="purple")
plt.title("Revenue with years")
plt.xlabel("Years")
plt.ylabel("Revenue")

for i in range(len(years)):
    plt.text(years[i], revenue[i], str(revenue[i]), ha="center")

plt.ylim(0,max(revenue)+100)
```

## Use of annotate

```python
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

plt.scatter(age,bp, color="red", s=bp, alpha=0.9)
plt.title("Scatter Ploting")
plt.xlabel("Age")
plt.ylabel("BP")
for i in range(len(people)):
    plt.annotate(people[i] ,xy=(age[i],bp[i]),xytext=(age[i]+2,bp[i]+1))

plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+15)
```

## ***Pie Charts***
- Use for cetagorical representation of  Data

```python
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
plt.style.use("dark_background")
plt.pie(amounts, labels=expenses)


plt.style.available
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

plt.pie(amounts, labels=expenses)

# set style
plt.style.use("ggplot")

# add %-age of each slice
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%')

#set colors
colors = ["purple", "green", "blue", "pink", "yellow"]
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', colors=colors)

# wedgeprops
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', 
        wedgeprops={"edgecolor":"black", "linewidth": 1, "linestyle":"-.", "fill":False}
       )
```

```python
# explode
explode = [0, 0, 0, 0.2, 0]
# plt.pie(amounts, labels=expenses, explode=explode)

# shadow
# plt.pie(amounts, labels=expenses, explode=explode, shadow=True)

# angle
plt.pie(amounts, labels=expenses, explode=explode, shadow=True, startangle=90) # top
plt.pie(amounts, labels=expenses, explode=explode, shadow=True, startangle=180) # flips
```
