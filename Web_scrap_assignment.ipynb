```python
links = set()
for a in soup.find_all('a', href=True):
    href = a['href']
    if '/story/' in href:
        # Agar href pehle se https:// se shuru ho toh seedha lo
        # Warna express.pk lagao aage
        if href.startswith('http'):
            full_link = href
        else:
            full_link = "https://express.pk" + href
        links.add(full_link)
```

```python
data=[]
```

```python
for link in links:
    r=requests.get(link)
    s=BeautifulSoup(r.text,'html.parser')
    h1=s.find('h1')
    title=h1.get_text() if h1 else "N/A"

     # Title
    h1 = s.find('h1')
    title = h1.get_text() if h1 else 'N/A'
    # Author
    author_tag = s.select_one('.author')
    author = author_tag.get_text() if author_tag else 'N/A'
    # Date — 3 jagah check karta hai
    date_tag = s.find('time')
    if date_tag:
        date = (date_tag.get('datetime') or
                date_tag.get('content') or
                date_tag.get_text())
    else:
        date = 'N/A'
    # Image — crash nahi hoga agar image na ho
    img = s.find('img')
    image = img['src'] if img else 'N/A'
    # Pehle 2 paragraphs
    paras = s.find_all('p')
    body = [p.get_text() for p in paras if len(p.get_text()) > 50][:2]
    data.append({
        'Title'  : title,
        'Link'   : link,
        'Image'  : image,
        'Author' : author,
        'Date'   : date,
        'Body'   : body,
    })
```

```python
df=pd.DataFrame(data)
df
```
