# 5. ***Conclusion***

The data shows that AI can either help or harm students — it depends on **how much** they use it.

- **Low to Moderate AI use (levels 1–5):** Students in this group have the **best GPA improvement** and their **skill retention stays high (~76)**. This means AI is genuinely helping them study better without replacing their own learning.

- **High AI use (levels 7–10):** Students here show a clear drop in outcomes:
  - GPA improvement becomes smaller
  - **Skill retention falls sharply (down to ~63–65)** — showing they are just getting answers instead of actually learning
  - **Exam anxiety increases**
  - **Burnout risk jumps drastically** (95% high burnout at max dependency vs only 10% at low dependency)

**Takeaway:** Used in moderation, AI acts like a helpful study assistant — it saves time and slightly boosts performance without hurting understanding. But when students over-rely on AI, it starts behaving like a shortcut for cheating — grades stop improving, real skills disappear, and stress/burnout go up.

So the answer to *"Is AI helping students or making them cheaters?"* is: **it's helpful in moderation, but harmful in excess.**
