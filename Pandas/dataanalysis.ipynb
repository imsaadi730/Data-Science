### Outlier Detection

```python
def findoutliers(data,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    IQR=q3-q1
    lb=q1-1.5*IQR
    ub=q3+1.5*IQR
    print(lb)
    print(ub)
    return df[(df[column]<lb) | (df[column]>ub)]
```

```python
q1=df["Fare"].quantile(0.25)
q3=df["Fare"].quantile(0.75)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
print(lb)
print(ub)
df[(df["Fare"]<lb) | (df["Fare"]>ub)]
```
