```python
# Basic Plot
plt.plot(df["Score"],df["Healthy life expectancy"])
plt.title("Healthy life expectancy")
plt.xlabel("Score")
plt.ylabel("Healthy life expectancy")
```

```python
sns.lineplot(x="Score",y="Healthy life expectancy", data=df)
```

```python
sns.lineplot(x="age",y="goals",data=Fifa)
```

```python
sns.scatterplot(x="age",y="goals",data=Fifa)
```

```python
Fifa["goals"]=Fifa["goals"]*100
```

Histogram: when we want to data distribution in a single column(only for 1 column)

```python
sns.histplot(x="goals",data=Fifa)
```
